### Upload data from previous lab as dataframe

```python
with open('canada_postal_codes.csv','r') as file:
    canada_df = pd.DataFrame(pd.read_csv(file, index_col=0))

canada_df
```

```python
with open('Geospatial_Coordinates.csv','r') as file:
    geo_df = pd.DataFrame(pd.read_csv(file, index_col=0)).reset_index()

geo_df
```

```python
# As there are different column names in canada_df and geo_df - we should make them the same
geo_df.rename(columns={geo_df.columns[0]: canada_df.columns[0]}, inplace=True)
geo_df
```

```python
# Merging canada_df with geo_df
geo_canada_df = canada_df.merge(geo_df)
geo_canada_df
```
